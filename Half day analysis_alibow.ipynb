```python
def average_over_draws(data):
    out = data
    out['mean'] = out.mean(axis=1)
    out['upper'] = out.quantile(0.975,axis=1)
    out['lower'] = out.quantile(0.025,axis=1)
    out = out[['mean','upper','lower']]
    out = out.reset_index()
    
    out['order'] = [2,4,6,8,1,3,5,7]    
    #out = out.append(out.where(out['order'] == 5).dropna())
    #out = out.append(out.where(out['order'] == 6).dropna())
    #out['order'] = [2,4,6,8,1,3,5,7,6.1,6.2] 
    out = out.sort_values(by='order')
    return out
```

```python
def average_over_draws(data):
    out = data
    out['mean'] = out.mean(axis=1)
    out['upper'] = out.quantile(0.975,axis=1)
    out['lower'] = out.quantile(0.025,axis=1)
    out = out[['mean','upper','lower']]
    out = out.reset_index()
    out = out.sort_values(by='sex', ascending=0)
    return out
```
