## RUN 5a BASELINE, MMN, BEP_CE_scale_up, BEP_TARGETED_scale_up

```python
#RUN 5a (bep_ce_universal & targeted)

master_cgf_5a = pd.DataFrame()

for i in list(range(len(output_dirs5a))):
    #cgf
    cgf_5a = pd.read_hdf(output_dirs5a[i] + 'cgf_z_scores.hdf').reset_index().sort_values(by=['input_draw','risk','measure','scenario','treatment_group','mother_status'])
    cgf_5a = cgf_5a.rename(columns={'value':'zscore'})
    cgf_5a['Z_SCORE'] = cgf_5a['zscore'] - 10
    
    #population
    pop = pd.read_hdf(output_dirs5a[i] + 'population.hdf').reset_index()
    pop = pop.where(pop['measure'] == 'total_population').dropna().rename(columns={'value':'start_population'}).drop(columns=['index','measure'])
    
    died = pd.read_hdf(output_dirs5a[i] + 'deaths.hdf')
    died = died.groupby(['input_draw', 'scenario','mother_status','treatment_group']).sum().rename(columns={'value':'deaths'}).reset_index()
    
    pop = pop.merge(died, right_on=['scenario','treatment_group','mother_status','input_draw'], left_on=['scenario','treatment_group','mother_status','input_draw'])
    
    data = pd.merge(cgf_5a, pop, left_on=['input_draw','scenario','treatment_group','mother_status'], right_on=['input_draw','scenario','treatment_group','mother_status'], how='outer')
   
    #print(data)
    data_1 = data.groupby(['risk','measure','scenario','treatment_group','mother_status','timepoint']).mean().drop(columns=['input_draw','index']).reset_index() #average over input draws
    data_2 = data.groupby(['risk','measure','scenario','treatment_group','mother_status','timepoint']).Z_SCORE.describe(percentiles=[.025, 0.975]).filter(['mean', '2.5%', '97.5%']).reset_index() #get 95%CIs
    
    data = pd.merge(data_1, data_2, left_on=['risk','measure','scenario','treatment_group','mother_status','timepoint'], right_on=['risk','measure','scenario','treatment_group','mother_status','timepoint'], how='outer')
    
    #print(data)
    
    data['alive'] = data['start_population'] - data['deaths']
    data['alive_x_cgf'] = data['Z_SCORE'] * data['alive'] #cannot use this as weight because the population is the population supplemented (or at birth) whereas the z-score is among alive population at 2 years
    data['location'] = locations[i]
    master_cgf_5a = master_cgf_5a.append(data)
    
master_cgf_5a.head()
```

```python
#RUN 5b (bep_ce_universal & targeted)

master_cgf_5b = pd.DataFrame()

for i in list(range(len(output_dirs5b))):
    #cgf
    cgf_5b = pd.read_hdf(output_dirs5b[i] + 'cgf_z_scores.hdf').reset_index().sort_values(by=['input_draw','risk','measure','scenario','treatment_group','mother_status'])
    cgf_5b = cgf_5b.rename(columns={'value':'zscore'})
    cgf_5b['Z_SCORE'] = cgf_5b['zscore'] - 10
    
    #population
    pop = pd.read_hdf(output_dirs5b[i] + 'population.hdf').reset_index()
    pop = pop.where(pop['measure'] == 'total_population').dropna().rename(columns={'value':'start_population'}).drop(columns=['index','measure'])
    
    died = pd.read_hdf(output_dirs5b[i] + 'deaths.hdf')
    died = died.groupby(['input_draw', 'scenario','mother_status','treatment_group']).sum().rename(columns={'value':'deaths'}).reset_index()
    
    pop = pop.merge(died, right_on=['scenario','treatment_group','mother_status','input_draw'], left_on=['scenario','treatment_group','mother_status','input_draw'])
    
    data = pd.merge(cgf_5b, pop, left_on=['input_draw','scenario','treatment_group','mother_status'], right_on=['input_draw','scenario','treatment_group','mother_status'], how='outer')
    
    data = data.groupby(['risk','measure','scenario','treatment_group','mother_status','timepoint']).mean().drop(columns=['input_draw','index']).reset_index() #average over input draws
    
    data['alive'] = data['start_population'] - data['deaths']
    data['alive_x_cgf'] = data['Z_SCORE'] * data['alive'] #cannot use this as weight because the population is the population supplemented (or at birth) whereas the z-score is among alive population at 2 years
    data['location'] = locations[i]
    master_cgf_5b = master_cgf_5b.append(data)
    
master_cgf_5b.head()
```

```python
master_stunting = pd.DataFrame()
master_wasting = pd.DataFrame()

for i in list(range(len(output_dirs))):
    cgf = pd.read_hdf(output_dirs[i] + 'category_counts.hdf').reset_index()
    cgf = (cgf.where(cgf['sex'] == 'male').dropna()
           .groupby(['input_draw','cgf_risk','risk_category','malnourishment_category']).sum()
           .reset_index())
    
    pop = (cgf.where(cgf['cgf_risk'] == 'stunting').dropna()
          .groupby(['input_draw','malnourishment_category'], as_index=False).sum()
          .rename(columns={'category_counts':'population'}))
    
    cgf = cgf.where(cgf['risk_category'] == ('cat1' or 'cat2'))
    
    data = cgf.merge(pop, right_on=['input_draw','malnourishment_category'], 
                     left_on=['input_draw','malnourishment_category'])
    stratified = data.groupby(['input_draw','malnourishment_category','cgf_risk']).sum().reset_index()
    
    overall = data.groupby(['input_draw','cgf_risk']).sum().reset_index()
    
    overall['malnourishment_category'] = 'all'
    data = stratified.append(overall)  
    
    stunted = (data.where(data['cgf_risk'] == 'stunting').dropna())
    stunted['percent'] = stunted['category_counts'] / stunted['population']
    
    #get 95%CI
    #stunted = (stunted.groupby(['malnourishment_category']).mean().drop(columns=['input_draw','category_counts','population']).reset_index()) 
    stunted = stunted.groupby(['malnourishment_category']).percent.describe(percentiles=[.025, 0.975]).filter(['mean', '2.5%', '97.5%']).reset_index() #get 95%CIs
    
    stunted['location'] = locations[i]
    master_stunting = master_stunting.append(stunted)   
    
    wasted = (data.where(data['cgf_risk'] == 'wasting').dropna())
    wasted['percent'] = wasted['category_counts'] / wasted['population']
    
    #get 95%CI
    #wasted = (wasted.groupby(['malnourishment_category']).mean().drop(columns=['input_draw','category_counts','population']).reset_index()) 
    wasted = wasted.groupby(['malnourishment_category']).percent.describe(percentiles=[.025, 0.975]).filter(['mean', '2.5%', '97.5%']).reset_index() #get 95%CIs
    
    wasted['location'] = locations[i]
    master_wasting = master_wasting.append(wasted)
    master_wasting
```
