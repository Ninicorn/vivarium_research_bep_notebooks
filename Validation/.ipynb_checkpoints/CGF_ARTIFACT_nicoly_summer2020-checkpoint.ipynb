## #sex ratio

```python
# get population birth sex ratio from artifact data

locations = ['india', 'pakistan', 'mali', 'tanzania']

def load_artifact_sr(age_end):
    age_end = float(age_end)
    tab = pd.DataFrame()
    for i in locations:
        art = Artifact('/share/costeffectiveness/artifacts/vivarium_gates_bep/' + i + '.hdf', filter_terms=['draw == 0', 'year_start == 2017', 'age_start >=  0.076712', f'age_end = {age_end}'])
        print(i)
        sr = art.load('population.structure').reset_index()
        sr = sr.rename(columns={'value':'pop'})
        sr['total_pop'] = sr['pop'][0] + sr['pop'][1]
        sr['sex_ratio'] = sr['pop'] / sr['total_pop']
        tab=tab.append(sr)      
        
    return tab
```

## #stunting z-scores 

```python
# get population stunting z-scores from artifact data

locations = ['india', 'pakistan', 'mali', 'tanzania']

def load_artifact_stunting(age_end):
    tab = pd.DataFrame()
    for i in locations:
        art = Artifact('/share/costeffectiveness/artifacts/vivarium_gates_bep/' + i + '.hdf', filter_terms=['draw == 0', 'year_start == 2017', 'age_start >=  0.076712', f'age_end = {age_end}'])
        print(i)
        stunting = art.load('alternative_risk_factor.child_stunting.exposure').reset_index()
        stunting = stunting.rename(columns={'draw_0':'zscore'})
        stunting['z_score'] = stunting['zscore'] - 10
        #print(stunting)
        tab=tab.append(stunting)
    return tab

load_artifact_stunting(1)
```

## #any cgf z-scores

```python

locations = ['india', 'pakistan', 'mali', 'tanzania']

def load_artifact_cgf(which_cgf, age_end):
    tab = pd.DataFrame()
    for i in locations:
        art = Artifact('/share/costeffectiveness/artifacts/vivarium_gates_bep/' + i + '.hdf', filter_terms=['draw == 0', 'year_start == 2017', 'age_start >=  0.076712', f'age_end = {age_end}'])
        print(i)
        cgf = art.load('alternative_risk_factor.child_'+which_cgf+'.exposure').reset_index()
        cgf = cgf.rename(columns={'draw_0':'zscore'})
        cgf['z_score'] = cgf['zscore'] - 10
        tab=tab.append(cgf)
    return tab

#load_artifact_cgf('stunting', 5)
```

```python
def sr_mer(cgf_df, sr_df):
    mer = pd.merge(cgf_df[['location', 'sex', 'age_start', 'age_end', 'year_start', 'year_end', 'z_score']], sr_df[['location', 'sex', 'age_start', 'age_end', 'year_start', 'year_end', 'sex_ratio']], how='inner',on=['location', 'sex', 'age_start', 'age_end', 'year_start', 'year_end'])
    mer['z_score_weight']=mer['z_score']*mer['sex_ratio']   # make weighted z score by sex ratio
    return mer

def get_weighted_zscore(which_cgf, age_end):
    cgf = load_artifact_cgf(which_cgf, age_end)
    sr = load_artifact_sr(age_end)
    cgf_merged = sr_mer(cgf, sr)
    return cgf_merged

def get_overall_zscore(which_cgf, age_end):
    cgf = load_artifact_cgf(which_cgf, age_end)
    sr = load_artifact_sr(age_end)
    cgf_merged = sr_mer(cgf, sr)
    cgf_overall = cgf_merged.groupby(['location']).z_score_weight.sum().reset_index() # sum sex weighted z scores for overall value
    cgf_overall.rename(columns={"z_score_weight": "z_score_"+str(age_end)}, inplace=True)
    return cgf_overall
```
