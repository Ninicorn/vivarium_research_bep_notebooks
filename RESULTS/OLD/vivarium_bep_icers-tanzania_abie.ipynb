```python
df = a.reset_index().groupby(['input_draw_number', 'maternal_supplementation.scenario']).sum()
df
```

```python
cols = list(df.columns)
[c for c in cols if ('treatment_bep' in c and 'birth' in c)]
```

```python
total_dalys = pd.Series(0, index=df.index)
for metric in ['ylls', 'ylds']:
    for cause in ['protein_energy_malnutrition', 'measles',
                  'diarrheal_diseases', 'lower_respiratory_infections',
                  'neonatal_disorders',
                  'meningitis', 'other_causes']:
        if metric == 'ylds' and cause in ['neonatal_disorders', 'other_causes']:
            continue
        for age in ['early_neonatal', 'late_neonatal', '1mo_to_6mo', '6mo_to_1', '1_to_4']:
            for maternal_bmi in ['normal', 'malnourished']:
                for tx in ['ifa', 'bep', 'mmn', 'none']:
                    col = f'{metric}_due_to_{cause}_in_age_group_{age}'
                    col += f'_mother_{maternal_bmi}_treatment_{tx}'
                    total_dalys += df[col]
```
